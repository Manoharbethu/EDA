### Type of distribution does the calories column in the food dataset follow, & Coverting it to a normal distribution?

```python
# Column to analyze
column = 'Calories'

# Check for normality
shapiro_test = shapiro(df[column])

# Print the results
print('Shapiro-Wilk test:', shapiro_test)

# If the data is not normally distributed, we can convert it to a normal distribution using the log-transformation.
# To do this, we can use the following code:

df = df[df['Calories'] > 0]
#Removes all rows from the DataFrame where the Calories column is equal to 0. 
#This is necessary because the log-transformation is undefined at 0

df.loc[df['Calories'] == 0, 'Calories'] = 1e-6
#Replaces all 0 values in the Calories column with a very small positive value, 1e-6. 
#This allows us to log-transform the data without encountering any errors

#Creates a new column in the DataFrame called log_Calories that contains the log-transformed values of the Calories column.
df['log_Calories'] = np.log(df['Calories'])

# Shapiro-Wilk test on the log_Calories column.
shapiro_test = shapiro(df['log_Calories'])

print('Shapiro-Wilk test after log-transformation:', shapiro_test)
```

## 7)Hypothesis Testing
If you have specific hypotheses about the relationship between two or more variables in your dataset, you can use hypothesis testing to test the hypotheses.


## T Test
A t-test is a type of inferential statistic which is used to determine if there is a significant difference between the means of two groups which may be related in certain features

T-test has 2 types : 1. one sampled t-test 2. two-sampled t-test.

A one-sampled t-test is used to compare the mean of a sample to a known population mean.

A two-sampled t-test is used to compare the means of two independent samples.

```python
from scipy.stats import ttest_ind
#ttest_ind is a function in the scipy.stats library that performs a two-sampled t-test

# Select the data for the two groups
high_fat_group = df[df['Total Fat'] > df['Total Fat'].mean()]
low_fat_group = df[df['Total Fat'] <= df['Total Fat'].mean()]

# Convert the data to NumPy arrays
high_fat_group_calories = np.array(high_fat_group['Calories'])
low_fat_group_calories = np.array(low_fat_group['Calories'])

# Perform the t-test
t_statistic, p_value = ttest_ind(high_fat_group_calories, low_fat_group_calories)

# Print the results
print('t-statistic:', t_statistic)
print('p-value:', p_value)

if p_value < 0.05:    # alpha value is 0.05 or 5%
    print("we are rejecting null hypothesis")
else:
    print("we are accepting null hypothesis")
```

###### 
Since the p-value is very small, we can reject the null hypothesis

The mean calorie intake of foods with high fat content is significantly greater than the mean calorie intake of foods with low fat content.

The t-test you performed in the code snippet is a two-sampled t-test. This is because you are comparing the means of two independent groups: the high-fat group and the low-fat group.

This information can be used to make more informed decisions about diet. For example, if you are trying to lose weight, you may want to avoid foods with high fat content.

## ANOVA Test(F-Test)
The t-test works well when dealing with two groups, but sometimes we want to compare more than two groups at the same time.


For example, if we wanted to test whether petal_width age differs based on some categorical variable like species, we have to compare the means of each level or group the variable

### One Way F-test(Anova) :-
It tell whether two or more groups are similar or not based on their mean similarity and f-score.

Example : there are 3 different category of iris flowers and their petal width and need to check whether all 3 group are similar or not 

```python
from scipy.stats import f_oneway
#f_oneway is a function in the scipy.stats library that performs a one-way ANOVA test.

# Get the food types
food_types = df['Proteins'].unique()

# Convert the data to NumPy arrays
calorie_data = np.array(df['Calories'])

# Perform the ANOVA test
f_statistic, p_value = f_oneway(calorie_data, food_types)

# Print the results
print('F-statistic:', f_statistic)
print('p-value:', p_value)

if p_value<0.05:
    print("we reject null hypothesis")
else:
    print("we accept null hypothesis")
```

####
Since the p-value is very small, we can reject the null hypothesis

This means that the mean calorie intake of some protein levels is significantly different from the mean calorie intake of other protein levels.

This information can be used to make more informed decisions about diet. For example, if you are trying to lose weight, you may want to avoid foods with high protein levels, as they may also have a high mean calorie intake.

## Chi-Square Test- 

The test is applied when you have two categorical variables from a single population. It is used to determine whether there is a significant association between the two variables.

```python
from scipy.stats import chi2_contingency
#chi2_contingency function in the scipy.stats library performs a chi-squared test of independence

# Create a contingency table
contingency_table = pd.crosstab(df['Item'], df['Calories'])

# Calculate the chi-squared statistic and p-value
chi2_statistic, p_value, dof, expected_values = chi2_contingency(contingency_table)

print('Chi-squared statistic:', chi2_statistic)
print('Degrees of freedom:', dof)
print('p-value:', p_value)

if p_value < 0.05:
    print('we reject null hypothesis')
else:
    print('we accept null hypothesis')
```

##### 
Since the chi_square is very small we can reject the null hypothesis

This means that the distribution of calories is significantly different for different items, Or we can say, some items have a significantly different mean calorie intake than other items.

This information can be used to make more informed decisions about diet. For example, if you are trying to lose weight, you may want to avoid items with a high mean calorie intake.
